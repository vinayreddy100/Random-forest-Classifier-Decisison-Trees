```python
df_alexa = df_alexa.drop(['date', 'rating'], axis =1)
```

```python
#use dummies to convert variation
```

```python
variation_dummies = pd.get_dummies(df_alexa['variation'], drop_first = True)
```

```python
variation_dummies
```

```python
df_alexa.drop(['variation'], axis = 1, inplace = True)
```

```python
df_alexa = pd.concat([df_alexa, variation_dummies], axis =1)
```

```python
df_alexa
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
alexa_countvectorizer = vectorizer.fit_transform(df_alexa['verified_reviews'])
```

```python
alexa_countvectorizer.shape
```

```python
print(vectorizer.get_feature_names())
```

```python
print(alexa_countvectorizer.toarray())
```

```python
df_alexa.drop(['verified_reviews'], axis = 1, inplace = True)
```

```python
df_alexa
```

```python
# encoded reviews
```

```python
encoded_reviews = pd.DataFrame(alexa_countvectorizer.toarray())
```

```python
df_alexa = pd.concat([df_alexa, encoded_reviews], axis = 1)
```

```python
df_alexa
```

```python
X = df_alexa.drop(['feedback'], axis = 1)
```

```python
Y = df_alexa['feedback']
```

```python
X
```

```python
Y
```

```python
X.shape
```

```python
# model 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
# lest train our classifier
```

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
randomforest_classifier = RandomForestClassifier(n_estimators = 150, criterion = 'entropy')
```

```python
randomforest_classifier.fit(X_train, Y_train)
```

```python
## evaluating the model
```

```python
y_predict_train = randomforest_classifier.predict(X_train)
```

```python
y_predict_train
```

```python
cm = confusion_matrix(Y_train, y_predict_train)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(Y_train, y_predict_train))
```

```python
y_predict_test = randomforest_classifier.predict(X_test)
```

```python
y_predict_test
```

```python
cm = confusion_matrix(Y_test, y_predict_test)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(Y_test, y_predict_test))
```

```python
df_alexa = pd.read_csv('amazon_alexa.tsv', sep ='\t')
```

```python
df_alexa = pd.concat([df_alexa, pd.DataFrame(alexa_countvectorizer.toarray())], axis = 1)
```

```python
X = df_alexa.drop(['date', 'rating', 'variation', 'verified_reviews', 'feedback'], axis = 1)
```

```python
X
```

```python
Y = df_alexa['feedback']
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 5)
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
randomforest_classifier = RandomForestClassifier(n_estimators = 150, criterion = 'entropy')
randomforest_classifier.fit(X_train, Y_train)
y_predict = randomforest_classifier.predict(X_test)
cm = confusion_matrix(Y_test, y_predict)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(Y_test, y_predict))
```
